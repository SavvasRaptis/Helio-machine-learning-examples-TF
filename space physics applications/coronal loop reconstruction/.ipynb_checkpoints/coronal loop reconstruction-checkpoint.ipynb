## Importing libraries and data files

```python
import os
import tensorflow as tf
import datetime
import sys
from os import path
import numpy as np
import pandas as pd
from tensorflow import keras 
import xgboost as xgb
#data to vectors
# Please note we have we have 5000 loops with 1500 points each
#2d
X2D = np.load("X2D.npy")
Y2D = np.load("Y2D.npy")
LNGTH_L2D = np.load("LNGTH_L2D.npy")
DST2D_FP = np.load("DST2D_FP.npy")
angle_top = np.load("angle_top.npy")

#3d
Z3D = np.load("Z3D.npy")
```

## Make test and train set



```python
data2D_training = np.zeros((3750, 1500, 5))
data2D_testing = np.zeros((1250, 1500, 5))
data3D_training = np.zeros((3750, 1500))
data3D_testing = np.zeros((1250, 1500))

#print(data2.shape)

for i in range(3750): 
  data2D_training[i,:,0] = X2D[:,i]
  data2D_training[i,:,1] = Y2D[:,i]
  data2D_training[i,:,2] = LNGTH_L2D[i]
  data2D_training[i,:,3] = DST2D_FP[i]
  data2D_training[i,:,4] = angle_top[i]

for i in range(1250): 
  data2D_testing[i,:,0] = X2D[:,i]
  data2D_testing[i,:,1] = Y2D[:,i]
  data2D_testing[i,:,2] = LNGTH_L2D[i]
  data2D_testing[i,:,3] = DST2D_FP[i]
  data2D_testing[i,:,4] = angle_top[i]


for i in range (3750):
  data3D_training[i,:1500] = Z3D[:,i]

for i in range (1250):
  data3D_testing[i,:1500] = Z3D[:,i]


print(data3D_training.shape)
print(data3D_testing.shape)
print(data2D_training.shape)
print(data2D_testing.shape)
```

## Normalizing the dataset 

```python
from sklearn import preprocessing

# Add a min_max scaler which is fited in training data
scaler_inputs_training_0 = preprocessing.StandardScaler() 
scaler_inputs_training_1 = preprocessing.StandardScaler()   
scaler_inputs_training_2 = preprocessing.StandardScaler() 
scaler_inputs_training_3 = preprocessing.StandardScaler() 
scaler_inputs_training_4 = preprocessing.StandardScaler() 
scaler_outputs_training = preprocessing.StandardScaler()

data2D_training[:,:,0] = scaler_inputs_training_0.fit_transform(data2D_training[:,:,0]) 
data2D_training[:,:,1] = scaler_inputs_training_1.fit_transform(data2D_training[:,:,1]) 
data2D_training[:,:,2] = scaler_inputs_training_2.fit_transform(data2D_training[:,:,2]) 
data2D_training[:,:,3] = scaler_inputs_training_3.fit_transform(data2D_training[:,:,3]) 
data2D_training[:,:,4] = scaler_inputs_training_4.fit_transform(data2D_training[:,:,4]) 
data3D_training = scaler_outputs_training.fit_transform(data3D_training)


data2D_testing[:,:,0] = scaler_inputs_training_0.transform(data2D_testing[:,:,0]) 
data2D_testing[:,:,1] = scaler_inputs_training_1.transform(data2D_testing[:,:,1]) 
data2D_testing[:,:,2] = scaler_inputs_training_2.transform(data2D_testing[:,:,2]) 
data2D_testing[:,:,3] = scaler_inputs_training_3.transform(data2D_testing[:,:,3]) 
data2D_testing[:,:,4] = scaler_inputs_training_4.transform(data2D_testing[:,:,4]) 
data3D_testing = scaler_outputs_training.transform(data3D_testing)



X_train = data2D_training
X_test = data2D_testing
y_train = data3D_training
y_test = data3D_testing
```

# Visualziating results (simple metrics)

Let's see the results

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import sklearn.metrics as metrics

#predict loops in the test set
predictions = model_used.predict(X_test[:,:])
predictions_denormalized = scaler_outputs_training.inverse_transform(predictions)
y_test_denormalized=scaler_outputs_training.inverse_transform(y_test)
    
def regression_metrics(y_true, y_predictions):

    explained_variance=metrics.explained_variance_score(y_true, y_predictions)
    mean_absolute_error=metrics.mean_absolute_error(y_true, y_predictions) 
    mse=metrics.mean_squared_error(y_true, y_predictions) 
    median_absolute_error=metrics.median_absolute_error(y_true, y_predictions)
    r2=metrics.r2_score(y_true, y_predictions)

    print('explained_variance: ', round(explained_variance,3))    
    print('median absolute error: ', round(median_absolute_error,3))    
    print('r2: ', round(r2,3))
    print('MAE: ', round(mean_absolute_error,3))
    print('MSE: ', round(mse,3))
    print('RMSE: ', round(np.sqrt(mse),3))
    
regression_metrics(y_test_denormalized,predictions_denormalized)
```
